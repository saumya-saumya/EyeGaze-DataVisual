```python
ggazedf = pd.read_csv('GraphGaze.csv',names = ['Time', 'ID', 'ScreenL X', 'ScreenL Y', 'CamL X', 'CamL Y', 'DistL', 'PupilL', 'CodeL',
                                                'ScreenR X', 'ScreenR Y', 'CamR X', 'CamR Y', 'DistR', 'PupilR', 'CodeR'])
```

```python
ggazedf.head()
```

```python
ggazedf = ggazedf[(ggazedf['CodeL'] == 0) & (ggazedf['CodeR'] == 0)]
```

```python
ggazedf.head()
```

```python
ggazedf['Screen X'] = ggazedf[['ScreenL X','ScreenR X']].mean(axis = 1)
```

```python
ggazedf['Screen Y'] = ggazedf[['ScreenL Y','ScreenR Y']].mean(axis = 1)
```

```python
ggazedf['Pupil'] = ggazedf[['PupilL','PupilR']].mean(axis = 1)
```

```python
ggazedf['Dist'] = ggazedf[['DistL','DistR']].mean(axis = 1)
```

```python
ggazedf.drop(['ScreenL X', 'ScreenL Y', 'ScreenR X', 'ScreenR Y', 'CamL X', 'CamL Y', 'CodeL', 'CodeR', 'CamR X', 'CamR Y', 'PupilL', 'PupilR', 'DistL', 'DistR'], axis = 1, inplace = True)
```

```python
ggazedf.head()
```

```python
gfixedgaze = gfixdf.merge(ggazedf, on = ['Time'], how='left').drop(['ID_x', 'ID_y'], axis = 1)
```

```python
for val in gfixedgaze['Pupil']:
    if 2.95 <= val < 3.12:
        gfixedgaze.loc[gfixedgaze['Pupil'] == val,'PupilDilation'] = 'A'
    elif 3.12 <= val < 3.29:
#         print('B',val)
        gfixedgaze.loc[gfixedgaze['Pupil'] == val,'PupilDilation'] = 'B'
    elif 3.29 <= val < 3.49:
        gfixedgaze.loc[gfixedgaze['Pupil'] == val,'PupilDilation'] = 'C'
    else:
        gfixedgaze.loc[gfixedgaze['Pupil'] == val,'PupilDilation'] = 'D'
```

```python
gfixedgaze['Minutes'] = gfixedgaze['Time'].apply(lambda x:x/60000)
```

```python
import plotly.graph_objects as go
import pandas as pd

minute=3
minutes = ["3", "6", "9", "12", "15", "18", "21", "24", "27"]

# make list of Dilation
dilations = list(gfixedgaze['PupilDilation'].unique())
dilations

# dilations = []
# for dilation in gfixedgaze["PupilDilation"]:
#     if dilation not in dilations:
#         dilations.append(dilation)

# print(dilations)
# make figure
fig_dict = {
    "data": [],
    "layout": {},
    "frames": []
}

# fill in most of layout
fig_dict["layout"]["xaxis"] = {"range": [0, 1300], "title": "Screen X",'gridcolor': '#FFFFFF'}
fig_dict["layout"]["yaxis"] = {"range": [0, 1300], "title": "Screen Y",'gridcolor': '#FFFFFF'}
fig_dict["layout"]["hovermode"] = "closest"

fig_dict["layout"]["sliders"] = {
    "args": [
        "transition", {
            "duration": 1000,
            "easing": "cubic-in-out"
        }
    ],
    "initialValue": "3",
    "plotlycommand": "animate",
    "values": minutes,
    "visible": True
}

sliders_dict = {
    "active": 10,
    "yanchor": "top",
    "xanchor": "left",
    "currentvalue": {
        "font": {"size": 20},
        "prefix": "Minute:",
        "visible": True,
        "xanchor": "right"
    },
    "transition": {"duration": 300, "easing": "cubic-in-out"},
    "pad": {"b": 10, "t": 50},
    "len": 0.9,
    "x": 0.1,
    "y": 0,
    "steps": []
}
fig_dict["layout"]["updatemenus"] = [
    {
        "buttons": [
            {
                "args": [None, {"frame": {"duration": 500, "redraw": False},
                                "fromcurrent": True, "transition": {"duration": 300,
                                                                    "easing": "quadratic-in-out"}}],
                "label": "Play",
                "method": "animate"
            },
            {
                "args": [[None], {"frame": {"duration": 0, "redraw": False},
                                  "mode": "immediate",
                                  "transition": {"duration": 0}}],
                "label": "Pause",
                "method": "animate"
            }
        ],
        "direction": "left",
        "pad": {"r": 10, "t": 87},
        "showactive": False,
        "type": "buttons",
        "x": 0.1,
        "xanchor": "right",
        "y": 0,
        "yanchor": "top"
    }
]


# make data
minute = 3
for dilation in dilations:
    gfixedgaze_by_minute = gfixedgaze[gfixedgaze["Minutes"] <= minute]
    gfixedgaze_by_minute_and_dil = gfixedgaze_by_minute[gfixedgaze["PupilDilation"] == dilation]

    data_dict = {
        "x": list(gfixedgaze_by_minute_and_dil["X"]),
        "y": list(gfixedgaze_by_minute_and_dil["Y"]),
        "mode": "markers",
        "text": list(gfixedgaze_by_minute_and_dil["PupilDilation"]),
        "marker": {
            "sizemode": "area",
            "sizeref":1,
            "size": list(gfixedgaze_by_minute_and_dil["Duration"])
        },
        "name": dilation
    }
    fig_dict["data"].append(data_dict)

# make frames
for minute in minutes:
    frame = {"data": [], "name": str(minute)}
    for dilation in dilations:
        gfixedgaze_by_minute = gfixedgaze[gfixedgaze["Minutes"] <= int(minute)]
        #print(gfixedgaze_by_minute)
        gfixedgaze_by_minute_and_dil = gfixedgaze_by_minute[
            gfixedgaze_by_minute["PupilDilation"] == dilation]

        data_dict = {
            "x": list(gfixedgaze_by_minute_and_dil["X"]),
            "y": list(gfixedgaze_by_minute_and_dil["Y"]),
            "mode": "markers",
            "text": list(gfixedgaze_by_minute_and_dil["PupilDilation"]),
            "marker": {
                "sizemode": "area",
                "sizeref": 2,
                "size": list(gfixedgaze_by_minute_and_dil["Duration"])
            },
            "name": dilation
        }
        frame["data"].append(data_dict)

    fig_dict["frames"].append(frame)
    slider_step = {"args": [
        [minute],
        {"frame": {"duration": 300, "redraw": False},
         "mode": "immediate",
         "transition": {"duration": 300}}
    ],
        "label": minute,
        "method": "animate"}
    sliders_dict["steps"].append(slider_step)


fig_dict["layout"]["sliders"] = [sliders_dict]

fig = go.Figure(fig_dict)

fig.show()
```
